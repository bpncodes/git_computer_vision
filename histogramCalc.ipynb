```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/bricks.jpg')
```

```python
colors = ('r','g','b')
```

```python
for i,col in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[255],[0,255])
    plt.plot(hist, color=col)
```
